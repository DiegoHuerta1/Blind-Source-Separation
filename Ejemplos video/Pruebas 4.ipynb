## Separar usando varios frames en cuadricula

```python
def construir_imagen_n_frames(frames_video, indices, n):
    """
    Dados muchos frames de un video e índices de n^2 frames,
    construir una matriz que se componga de n x n bloques en una cuadrícula.
    """
    if len(indices) != n * n:
        raise ValueError("El número de índices debe ser igual a n^2")
    
    # Construir la matriz de bloques
    matriz = [[frames_video[indices[i * n + j]] for j in range(n)] for i in range(n)]
    
    return np.block(matriz)
```

### OVFCA

```python
def esperanza_condicional_general(A, num_bloques):
    """
    Calcula la esperanza condicional de A, dividiéndola en una cuadrícula de num_bloques x num_bloques.
    """
    N = A.shape[0] // num_bloques  # Determinar tamaño de cada bloque
    indices = [slice(i * N, (i + 1) * N) for i in range(num_bloques)]  # Definir índices para los bloques
    # Calcular trazas de los bloques
    traces = np.array([[A[idx_i, idx_j].trace() for idx_j in indices] for idx_i in indices]) / N
    # Construir el resultado usando np.kron para replicar los valores con la identidad
    return np.kron(traces, np.eye(N))


def esperanza_condicional(A):
    return esperanza_condicional_general(A, 4)

# considerar la norma de frobenius en la imagen de la esperanza condicional
def frobenius_norm(A):
    return np.linalg.norm(A, 'fro')
```

```python
start_time = time.time()

# ejecutar OVFCA
ovfca = OVFCA_fast(señales_y_processed, esperanza_condicional_F = esperanza_condicional, norma_en_B = frobenius_norm)
ovfca.separar_señaes(repeticiones_optimizacion = 1, iniciar_W_ortogonal = True)
elapsed_time = time.time() - start_time

print(f"Ejecucion duro {elapsed_time} segundos")
```

```python
# ver el resultado
ovfca.ver(señales_x_originales = señales_originales_x_processed, figsize = (7, 12))
```

```python
# salvar la matriz estimada por este metodo
A_hat_ovfca_grid = ovfca.A_hat
A_hat_inv_ovfca_grid = ovfca.A_hat_inversa
```
